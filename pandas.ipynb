I'd like to do some preprocessing to the data before continuing the rest of the assignment  

```python
#Range the age from 1 to 4

heart.loc[(heart['age'] >= 29) & (heart['age'] <= 40), 'age'] = 1
heart.loc[(heart['age'] >= 41) & (heart['age'] <= 50), 'age'] = 2
heart.loc[(heart['age'] >= 51) & (heart['age'] <= 60), 'age'] = 3
heart.loc[(heart['age'] >= 61) & (heart['age'] <= 77), 'age'] = 4
```

```python
heart.head()
```

Discretized the continuous features 

New method is 'qcut' which discretize variable into equal-sized buckets based on rank or based on sample quantiles.
documnation: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.qcut.html

```python
heart['trestbps']= pd.qcut(heart.trestbps, q=4, labels=False)
heart['chol']= pd.qcut(heart.chol, q=4, labels=False)
heart['thalach']= pd.qcut(heart.thalach, q=5, labels=False)
heart['oldpeak']= pd.qcut(heart.oldpeak, q=3, labels=False)
```
